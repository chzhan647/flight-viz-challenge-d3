```python
from dotenv import load_dotenv
import os
from langchain import OpenAI

from langchain import OpenAI
from langchain import PromptTemplate
import os


```

```python
load_dotenv()

openai_api_key = os.getenv('OPENAI_API_KEY', 'YourKey')



llm = OpenAI(temperature=0, openai_api_key=openai_api_key)
```

```python



prompt = """
Please categorize the cause into 
Pilot error, 
Pilot communication with airport crew,
Passager error,
Airport error,
Weather, 
Mechanical failure,
Air traffic controller, 
Airline negligence,
Other,
Wildlife,
Other,
Undetermined
of the following text: 

The inadequate communication with ground crew personnel, which resulted in the airplane’s collision with a deice truck.


"""

output = llm(prompt)
print (output)
```

```python
template = """
Please categorize the cause into 
Pilot error, 
Pilot communication with airport crew,
Passager error,
Airport error,
Weather, 
Mechanical failure,
Air traffic controller, 
Airline negligence,
Other,
Wildlife,
Other,
Undetermined
of the following text: 

{essay}
"""
prompt = PromptTemplate(
    input_variables=["essay"],
    template=template
)
```

```python
summary_list = []
for essay in df[df['ProbableCause'].notna()]['ProbableCause'][0:10].tolist():
    summary_prompt = prompt.format(essay=essay)
    
    num_tokens = llm.get_num_tokens(summary_prompt)
    print (f"This prompt + essay has {num_tokens} tokens")
    
    summary = llm(summary_prompt)
    summary_list.append(summary)
    print (f"Summary: {summary.strip()}")
    print ("\n")
```
